## 2. The question: decision, action, cost of a wrong call

**Search question:** Which safe content and search signals are associated with visibility, clicks, engagement, or movement — and which of those associations are strong enough to guide editorial attention?

**Unit of analysis:** one row = one pseudonymized content page (starter CSV grain).

**Output:** a signal report (grouped comparisons, effect sizes / associations, charts) plus practical, careful recommendations — not an automatic publish decision.

**Decision this improves:** which levers an SEO or content lead should emphasize when prioritizing work across a large inventory.

**Who acts / action:** an editor or SEO strategist uses the report to focus effort (e.g. freshness, depth, CTR/snippet review, engagement) instead of guessing.

**Cost of a wrong call:** wasted rewrite hours on weak levers, or ignoring stronger *observed* associations and leaving traffic on the table.

**Why data / ML can help (and why this is not "just train a model"):** many signals move together across thousands of pages; a hand-written if-statement cannot score them fairly. Analysis can rank associations and spell out caveats. A model may appear later as a *tool* for summarizing associations (e.g. feature importance), but the product is decision-support evidence — not a claim that we found Google's ranking factors.

**One-paragraph frame:** For SEO/content leads deciding which levers to emphasize, we will build a signal report from the anonymized content starter set, measuring associations between observed page metrics and visibility/click outcomes. A wrong call costs wasted editorial time on weak levers. A plain rule isn't enough because signals interact and vary by content type. We will claim only observational / directional / decision-support results.

## 3. Quick look at the data (2-3 real numbers)

The cell below loads the starter CSV and prints a few **real** counts that back Ranking Signal Analysis:

1. **Inventory size** — enough pages (and clients) to compare signals.
2. **Outcome spread** — share of pages with any clicks in 90 days (visibility ≠ clicks).
3. **Movement mix** — share with `trend_direction == "down"`, plus one observational association example (`word_count` vs log impressions).

These numbers motivate the lane; they do **not** prove causality. Note: `avg_position == 0` means missing data (not rank zero); rate columns like `ctr` are ×100 percentages.

```python
from pathlib import Path

import numpy as np
import pandas as pd

# Resolve path whether the kernel cwd is repo root or work/notebooks/
candidates = [
    Path("data/raw/content_refresh_anonymized.csv"),
    Path("../../data/raw/content_refresh_anonymized.csv"),
]
csv_path = next(p for p in candidates if p.exists())

df = pd.read_csv(csv_path)

n_rows = len(df)
n_clients = df["client_id"].nunique()
click_share = (df["clicks_90d"] > 0).mean()
down_share = (df["trend_direction"] == "down").mean()
pos_missing = int((df["avg_position"] == 0).sum())

wc_mask = df["word_count"].notna()
corr_words_imps = (
    df.loc[wc_mask, "word_count"]
    .corr(np.log1p(df.loc[wc_mask, "impressions_90d"]))
)

print(f"loaded: {csv_path}")
print(f"rows={n_rows:,}  clients={n_clients}  columns={df.shape[1]}")
print(f"share with any clicks (90d)={click_share:.1%}")
print(f"share trend_direction=='down'={down_share:.1%}")
print(f"avg_position==0 (no data sentinel)={pos_missing:,}")
print(f"corr(word_count, log1p impressions)≈{corr_words_imps:.3f}  [observational]")
print()
print("trend_direction mix:")
print(df["trend_direction"].value_counts().to_string())
```

## 4. Careful words: what I can and can't claim

**What I can claim (careful language):**
- **Observed** associations between safe page metrics and visibility/click/engagement outcomes in this snapshot.
- **Directional** patterns (e.g. "pages with X tend to look like Y in this dataset").
- **Decision-support** recommendations for where an editor might look first — always with caveats.

**What I cannot claim:**
- Google ranking factors, algorithm secrets, or that we can "predict Google."
- That changing a signal **causes** better ranks or more clicks (no experiment / causal design here).
- That a refresh will recover traffic.
- AI citations, AI rankings, or AI search visibility from these fields.

This lane is observational analysis plus evidence-backed recommendations — not causal proof and not "train a model for its own sake."

```python
# Claim vocabulary used in this notebook (keep claims inside this set)
allowed = ["observed", "measured", "directional", "decision-support", "associational"]
forbidden = ["causes", "Google ranking factor", "predict Google", "proven algorithm"]
print("allowed claim words:", ", ".join(allowed))
print("avoid claiming:", ", ".join(forbidden))
```
